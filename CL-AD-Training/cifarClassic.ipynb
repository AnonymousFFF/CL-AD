```python
import os
import numpy as np
from torchvision import datasets, transforms
from PIL import Image

# 
transform = transforms.Compose([
    transforms.ToTensor(),
])

# CIFAR-10
train_dataset = datasets.CIFAR10(root='./cifar10', train=True, download=True, transform=transform)
test_dataset = datasets.CIFAR10(root='./cifar10', train=False, download=True, transform=transform)

# 
class_names = train_dataset.classes

# 
save_dir = './cifar10_sorted'
os.makedirs(save_dir, exist_ok=True)

# 
for dataset in [train_dataset, test_dataset]:
    for idx in range(len(dataset)):
        image, label = dataset[idx]
        # PIL
        image = transforms.ToPILImage()(image)
        # 
        class_name = class_names[label]
        # 
        class_dir = os.path.join(save_dir, f'cifar_{class_name}')
        os.makedirs(class_dir, exist_ok=True)
        # 
        image.save(os.path.join(class_dir, f'{idx}.png'))
        # 
        print(f'Saved image {idx} of class {class_name}')

print("")
```
